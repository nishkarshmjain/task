```python
df = pd.read_json("source.json", orient='columns')
```

```python
df1 = df.sort_values(by='priority')
```

```python
df1
```

```python
df_manager = df1.drop("watchers",axis =1)
df_manager = df_manager.sort_values(by='priority')
```

```python
df_watcher = df1.drop("managers",axis =1)
df_watcher = df_watcher.sort_values(by='priority')
```

```python
#df_manager['managers'] = df_manager['managers'].astype('str')
```

```python
df_watcher['watchers'] =  tuple(list(df_watcher['watchers'].astype('str')))
```

```python
df_manager['managers'] =  list(df_manager['managers'])

```

```python
op = df_manager.groupby('managers').apply(lambda x: x.to_json(orient='records'))
```

```python
op.to_json('managers.json')
```

```python
op
```

```python
op_watchers = df_watcher.groupby('watchers').apply(lambda x: x.to_json(orient='records'))
```

```python
op_watchers.to_json('watchers.json')
```
